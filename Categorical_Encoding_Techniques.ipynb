```python
import pandas as pd
import numpy as np
```

```python
my_dict1 = {'Temperature': ['Hot', 'Cold', 'Warm', 'Cold']}
```

```python
df_temp = pd.DataFrame(my_dict1)
```

```python
df_temp
```

One hot encoding


```python
#To convert categorical values into boolean values
pd.get_dummies(df_temp)
```

```python
#To convert categorical values into numerical values
pd.get_dummies(df_temp).astype(int)
```

```python
my_dict2 = {'Sec-A' : ['Alberto', 'Asalah', 'Vasanthi', 'Isaac'], 'Sec-B' : ['Sumaya', 'Margo', 'Josephenine', 'Andrew'], 'Subject-Code' : [1,2,3,4]}
```

```python
df_student = pd.DataFrame(my_dict2)
```

```python
df_student
```

```python
pd.get_dummies(df_student)
```

```python
pd.get_dummies(df_student).astype(int)
```

```python
pd.get_dummies(df_student,columns = ['Sec-A', 'Sec-B']).astype(int)

```

Label Encoding


```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df_temp = le.fit_transform(df_temp)
```

```python
df_temp
```

```python
df_student['Sec-A'] = le.fit_transform(df_student['Sec-A'])
```
