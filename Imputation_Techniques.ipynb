```python
import pandas as pd
import numpy as np
```

```python
my_dict = {'First' : [95,90,np.nan,100], 'Second' : [30,45,56,np.nan], 'Third' : [np.nan,34,42,76]}
```

```python
df = pd.DataFrame(my_dict)
```

```python
df
```

```python
#To get the total numbers of null values in each column
df.isna().sum()
```

Zero Imputation

```python
df1 = df.fillna(0)
```

```python
df1
```

Mean Imputation


```python
df.mean()
```

```python
df2 = df.fillna(df.mean())
```

```python
df2
```

Median Imputation

```python
df.median()
```

```python
df3 = df.fillna(df.median())
```

```python
df3
```

**Hot deck Imputation**

Imputation with previous values


```python
df4 = df.fillna(method = 'pad')
```

```python
df4
```

Imputation with next values

```python
df5 = df.fillna(method ='bfill')
```

```python
df5
```
